```python
pretty('Dow Jones daily data, bid prices: open, high, low, close, & volume')
api.get_history(instrument = 'US30_USD', 
				start = '2023-01-01',
			    end = '2023-02-01',
			    granularity = 'D',
			   price = 'B').head(5)
```

```python
pretty('Euro to US Dollar: daily data, bid prices')
api.get_history('EUR_USD', 
				'2023-01-01',
			    '2023-02-01', 
				'D', 'B').head(5).style.format(precision = 4)
```

```python
pretty('Euro to US Dollar: daily data, ask prices')
d(api.get_history('EUR_USD', 
				'2023-01-01',
			    '2023-02-01', 
				'D', 'A').head(5).style.format(precision = 4))

pretty('Euro to US Dollar: daily data, bid prices')
api.get_history('EUR_USD', 
				'2023-01-01',
			    '2023-02-01', 
				'D', 'B').head(5).style.format(precision = 4)
```
